```python
# 引入套件
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#讀取csv文件並顯示出字串
df = pd.read_csv('lei.csv')
print(df.to_string()) 

df.plot(kind = 'scatter', x = 'Year', y = 'Coincident Index')

plt.show()
```
