```python
import fitz
import jieba
```

```python
# 開啟PDF文件
pdf_path = 'C:\\Users\\YI-PEI\\Desktop\\PL\\HW5\\1.pdf'  # 設定 PDF 文件的路徑

def extract_text_from_pdf(pdf_path):

    pdf_text = ""

    # 創建 PdfReader 對象
    pdf_reader = fitz.open(pdf_path)
    
    #得到PDF的頁數
    num_pages = pdf_reader.page_count
    
    for page_num in range(num_pages):
        page = pdf_reader[page_num]
        text = page.extract_text()
        print(f'Page {page_number + 1}:\n{text}')

```

```python
#保存從 PDF 中提取的所有文本
pdf_text = ""

#將每一頁文本用 get_text() 提取出來，並連接到 pdf_text 字符串中
pdf_reader = fitz.open(pdf_path)
num_pages = pdf_reader.page_count
for page_num in range (num_pages):
    page = pdf_reader[page_num]
    pdf_text += page.get_text()
    

# 關閉 PDF 文件
pdf_reader.close()
```

```python
#將提取的文本另存到output_1.txt文件中
with open('output_1.txt', 'w', encoding='utf-8') as file:
    file.write(pdf_text)
    

    print(pdf_text)
```

```python
# 將提取到的文本進行分詞
tokenized_text = jieba.lcut(pdf_text, cut_all=True )

# 查看分詞結果
print(tokenized_text)
```

```python
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#列出我所關心的分詞
filtered_words = ["和平", "經濟", "年輕人", "教育" ]

#計算詞頻
word_freq = Counter(filtered_words)


# 使用支援中文的字型
font_path = "C:\Windows\Fonts\msyh.ttc"  # Microsoft YaHei

# 生成文字雲，指定字體路徑
wordcloud = WordCloud(width=800, height=400, background_color="white", font_path=font_path).generate(" ".join(filtered_words))

# 顯示文字雲
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# 提取詞和對應的頻率
words = list(word_freq.keys())
freqs = list(word_freq.values())

# 確保詞和詞頻的長度相同
N = 10
words = words[:N]
freqs = freqs[:N]

# 顯示前 N 個詞的詞頻條狀圖
plt.bar(range(len(words)), freqs, align="center")  # 修改這裡
plt.xticks(range(len(words)), words, rotation=45)
plt.xlabel('詞')
plt.ylabel('詞頻')
plt.title('前 10 個詞的詞頻條狀圖')
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib

# 設置默認字體
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
matplotlib.rcParams['axes.unicode_minus'] = False  # 解決負號顯示問題

# 計算詞頻
word_freq = Counter(filtered_words)
words = list(word_freq.keys())
freqs = list(word_freq.values())

# 繪製條狀圖
plt.bar(words, freqs)
plt.xlabel('詞彙')
plt.ylabel('頻率')
plt.title('分詞頻率條狀圖')
plt.show()

```
